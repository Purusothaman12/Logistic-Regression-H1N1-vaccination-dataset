x,y should be splitted into x_train_x_test,y_train and y_Test to create model using logistic regression and to evaluate the performance of the model

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=30)
```

StandardScaler removes the mean and scales each feature/variable to unit variance

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

CREATING MODEL

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

PREDICTING Y_VALUES USING x_test

```python
y_pred=model.predict(x_test)
y_pred
```

```python
df["h1n1_vaccine"].value_counts()
```

```python
np.unique(y_pred)
```

Evaluating the perfomanse of model

confusion matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

classification report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
